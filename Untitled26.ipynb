```python
import pandas as pd
from gtts import gTTS
from tqdm import tqdm
import os

# Load CSV
df = pd.read_csv("tgl_Latn_part2.csv")
os.makedirs("audio", exist_ok=True)

print(f"Total sentences: {len(df)}")

# Convert each sentence to MP3
failed = []
for _, row in tqdm(df.iterrows(), total=len(df), desc="Converting"):
    try:
        filename = f"audio/sentence_{int(row['id']):06d}.mp3"
        if not os.path.exists(filename):  # skip already done
            tts = gTTS(text=str(row['source']), lang='tl')  # 'tl' = Tagalog
            tts.save(filename)
    except Exception as e:
        failed.append({"id": row['id'], "error": str(e)})

print(f"\n✅ Done! Success: {len(df) - len(failed)} | Failed: {len(failed)}")
```
